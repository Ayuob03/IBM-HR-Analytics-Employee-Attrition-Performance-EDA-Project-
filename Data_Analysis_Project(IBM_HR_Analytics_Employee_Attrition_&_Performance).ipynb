# **Insights**

**Overall attrition rate**

```python
attrition_rate = df["Attrition"].value_counts(normalize=True)["Yes"] * 100
print(f"➡ Overall Attrition Rate: {attrition_rate:.2f}% of employees left the company.")
```

**Overtime impact**

```python
overtime_attrition = df[df["OverTime"]=="Yes"]["Attrition"].value_counts(normalize=True)["Yes"] * 100
no_overtime_attrition = df[df["OverTime"]=="No"]["Attrition"].value_counts(normalize=True)["Yes"] * 100
print(f"➡ Attrition among overtime workers: {overtime_attrition:.2f}%")
print(f"➡ Attrition among non-overtime workers: {no_overtime_attrition:.2f}%")
```

**Gender attrition**

```python
gender_attrition = df.groupby("Gender")["Attrition"].value_counts(normalize=True).unstack() * 100
print("\n➡ Attrition by Gender (%):")
print(gender_attrition.round(2))
```

**Department attrition**

```python
dept_attrition = df.groupby("Department")["Attrition"].value_counts(normalize=True).unstack() * 100
print("\n➡ Attrition by Department (%):")
print(dept_attrition.round(2))
```

**Job satisfaction vs attrition**

```python
satisfaction_attrition = df.groupby("JobSatisfaction")["Attrition"].value_counts(normalize=True).unstack() * 100
print("\n➡ Attrition by Job Satisfaction (%):")
print(satisfaction_attrition.round(2))
```

**Average age and income of attrition vs non-attrition**

```python
summary_stats = df.groupby("Attrition")[["Age","MonthlyIncome","YearsAtCompany"]].mean().round(2)
print("\n➡ Average Profile (Attrition vs Non-Attrition):")
print(summary_stats)
```
